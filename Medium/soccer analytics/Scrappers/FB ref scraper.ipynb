```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# All EPL Players

```python
data=requests.get('https://fbref.com/en/comps/9/stats/Premier-League-Stats#stats_standard').text
```

```python
data=data.replace('<!--','<br>')
data=data.replace('-->','<br>')
```

```python
soup=BeautifulSoup(data,'html.parser')
tables=soup.find_all('table')
```

```python
headers=[x.text.strip() for x in tables[2].thead.find_all('tr')[1].find_all('th')]
headers.extend(['slug','code'])
```

```python
data=[]
rows=tables[2].tbody.find_all('tr')
for row in rows:
    try:
        int(row.find('th').text)
        temp=[row.find('th').text]
        temp.extend([x.text for x in row.find_all('td')])
        temp.append(row.find_all('td')[0].a.get('href').split('/')[-1])
        temp.append(row.find_all('td')[0].a.get('href').split('/')[-2])
        data.append(temp)
    except:
        pass
```

```python
df=pd.DataFrame(columns=headers, data=data)
df
```

```python
df.to_csv('epl_players0910.csv', index=None)
```
